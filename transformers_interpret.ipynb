# Предложения по выделению основных конструкций из текста

Наша команда предлагает 2 варианта решения, которые помогут экспертам в разборе классификации текста посредством анализа основных конструкций:
1. Выделение конструкций на основе ручной разметки (обучение NER модели после разметки корпуса данных)
2. Выделение конструкций на основе предсказаний модели (начиная от eli5 до transformers-interpret)

Рассмотрим минусы и плюсы каждого подхода

NER-модели:
1. Требуют ручной разметки в большом количестве
2. Сами по себе модели довольно тяжеловесны - условная модель от DeepPavlov для NER будет весить от 1.6 гб

1. ОЧЕНЬ высокое качество распознавания - у сильнейше NER модели от DeepPavlov F1 = 98.9
2. Машинное определение в силу разметчки человеком будет "подражать" человеческому определению

Способ 2 - интерпертация результатов трансформера
Минусы данного метода:
1. Интерпретация будет со стороны модели, а не человека
2. Учитывая токенизатор трансформер-моделей (byte-pair encoding), отображение будет по важности токенов, а не целых слов

Плюсы данного метода:
1. Нет человеческих затрат по разметке

Возможные варианты решения проблемы:
1. lime, eli5, etc. - аналогично transformers-interpret, только некоторые либо не поддерживают transforer models, либо рабоают хуже
2. keywords extraction - отображает не релевантные полученной метке конструкции, а топ n-gram, описывающих данный текст лучше всего - проще говоря, **использование bert + CountVectorizer/TfIDf не является решением проблемы с точки зрения постановки проблематики**
